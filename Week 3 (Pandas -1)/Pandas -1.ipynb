# Missing Values:

Question: Are there any missing values in the dataset, and if so, how should they be handled for each indicator? 

```python
df.isnull().sum()
```

```python
#prevents the extreme values from disproportionately influencing the imputed values.
median_age = df['Age'].median()
df['Age'].fillna(median_age, inplace=True)
```

# 2. Data Types:

```python
df.dtypes
```

```python
df['Salary'] = df['Salary'].replace('[\$,]', '', regex=True).replace('k', '000', regex=True)
```

```python
df['Easy Apply']=df["Easy Apply"].replace('-1','FALSE')
```

```python
df.head()
```

```python
z_scores = (df[['Age', 'Salary', 'Rating']] - df[['Age', 'Salary', 'Rating']].mean()) / df[['Age', 'Salary', 'Rating']].std()

# Define a threshold for outlier detection (e.g., Z-score greater than 3 or less than -3)
outlier_threshold = 3

# Identify potential outliers
potential_outliers = (z_scores.abs() > outlier_threshold).any(axis=1)

# Display rows with potential outliers
print("Rows with potential outliers:")
print(df[potential_outliers])
```

```python
# Replace potential outliers with a specific value (e.g., the mean)
df_adjusted = df.copy()
df_adjusted.loc[potential_outliers, ['Age', 'Salary', 'Rating']] = df[['Age', 'Salary', 'Rating']].mean()
```

```python
df['Location'] = df['Location'].apply(lambda x: x.split(' ')[0])
df['Location'] = df['Location'].apply(lambda x: x.split(',')[0])
df['Location'] = df['Location'].replace('New', 'New York', regex=True)
```

```python
df['Established'] = df['Established'].replace(-1, df['Established'].median())
```

```python
df['Easy Apply'] = df['Easy Apply'].str.lower().replace({'true': True, 'false': False}).astype(bool)
```

```python
df.dtypes
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
df['Rating'].fillna(df['Rating'].median(), inplace=True)
```

```python
# Access the categories from the category dtype
categories = df['Location'].astype('category').cat.categories

# Print the mapping
for code, location in enumerate(categories):
    print(f"Code: {code}, Location: {location}")
df['Location'] = df['Location'].astype('category').cat.codes
```

```python
# ques 15 (normalized )
df['Normalized_Rating'] = (df['Rating'] - df['Rating'].min()) / (df['Rating'].max() - df['Rating'].min())
```
